```python
df = df.rename(columns={
    "Invoice": "InvoiceNo",
    "Price": "UnitPrice"
})
```

```python

df.columns = df.columns.str.lower()
```

```python
df['unitprice'] = (
    df['unitprice']
    .astype(str)
    .str.replace(',', '.', regex=False)
    .astype(float)
)
```

```python
df['invoicedate'] = pd.to_datetime(
    df['invoicedate'],
    format="%d/%m/%Y %H:%M",
    errors="coerce"
)
```

```python
df = df.drop_duplicates(
    subset=["invoiceno", "stockcode", "invoicedate"]
)
```
